#**Object-Oriented Programming**

Object-oriented programming (OOP) organizes code into objects, which are instances of classes. These objects can contain both data (attributes) and the functions (methods) that operate on that data. OOP emphasizes the concept of encapsulation, allowing data and methods to be bundled together and interact with each other, making it easier to model and solve complex problems in a more structured and modular way. It also promotes the principles of inheritance and polymorphism, which enable code reuse and flexibility in software design.

```python
class vehicle:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print('move!!')

class Car(vehicle):
  def move(self):
    print("Drive!")

class Boat(vehicle):
  def move(self):
    print("Sail!")

class Plane(vehicle):
  def move(self):
    print("Fly!")

car1 = Car("Ford", "Mustang")       #Create a Car class
boat1 = Boat("Ibiza", "Touring 20") #Create a Boat class
plane1 = Plane("Boeing", "747")     #Create a Plane class

for i in (car1,boat1,plane1):
  print(i.brand)
  print(i.model)
  i.move()
```

```python
class Animals:
  def __init__(self, name, habitat):
    self.name = name
    self.habitat = habitat

  def animals_detail(self):
    print("This is {self.name} and they live in {self.habitat}")


class Amphibians(Animals):
  def animals_detail(self):
    print('They can live in both land and water')

class Arriels(Animals):
  def animals_detail(self):
    print('They fly')

class Aquatics(Animals):
  def animals_detail(self):
    print('They live in water')


Amphibian = Amphibians('Crocodile','Land and Water') #I've created an object for this three child classes
Arriel = Arriels('Eagle','Air')
Aquatic = Aquatics('Whale', 'Water')

for k in (Amphibian, Arriel, Aquatic):
  print(k.name)
  print(k.habitat)
  k.animals_detail()


```

**Encapsulation**

Encapsulation is used to restrict access to methods and variables. In encapsulation, code and data are wrapped together within a single unit from being modified by accident. Encapsulation is a mechanism of wrapping the data (variables) and code acting on the data (methods) together as a single unit.

```python
#Encapsulation

class transport:

  def __init__(self, brand, model, year):
    self.brandofvehicle = brand
    self.modelofvehicle = model
    self.yearofmfg= year

  def print_details(self):
    print(self.brandofvehicle, self.modelofvehicle, self.yearofmfg)
```

```python
class flowers:

  def __init__(self, name, color):
    self.name = name
    self.color = color

  def flower_details(self):
    print(self.name, self.color)

class Rose(flowers):
  def __init__(self, name, color, bloom):
    super().__init__(self, name, color)
    self.bloom = bloom

  def flower_details(self):
    print('Rose is blooming with a lovely fragrance."')

class Sunflower(flowers):
  def __init__(self, name, color, bloom):
    super().__init__(self, name, color)
    self.bloom = bloom
  def flower_details(self):
        print("A sunflower is turning towards the sun.")

```
