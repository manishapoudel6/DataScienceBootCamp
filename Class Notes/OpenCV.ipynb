```python
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

# cv2.imread used for image reading with orignal colour
image = cv2.imread('0_0_480_1_70__News_iStock-1218489833.jpg',1)
# 0 - 255 range for each pixel value
print(image.shape) #(h,w,c)
print(image[0][0][0]) # how 0,0 pixel looks like for 1 channel(color)
#cv2.imshow("image", img)
cv2_imshow(image) # In output 3 is channel which means color (Red, Green, Blue)
```
