https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html

how{‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’}, default ‘inner’ Type of merge to be performed.
left: use only keys from left frame, similar to a SQL left outer join; preserve key order.
right: use only keys from right frame, similar to a SQL right outer join; preserve key order.
outer: use union of keys from both frames, similar to a SQL full outer join; sort keys lexicographically.
inner: use intersection of keys from both frames, similar to a SQL inner join; preserve the order of the left keys.
cross: creates the cartesian product from both frames, preserves the order of the left keys.

```python
final_df = pd.merge(df_zomato, df_country, on = 'Country Code' , how = 'left')
```

```python
final_df
```

```python
country_names=final_df.Country.value_counts().index
country_val=final_df.Country.value_counts().values
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%', explode=(0.12, 0.15, 0))
#explode function it moves the output like pop out a bit outside
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

#**Python Lambda Functions**
An anonymous function that can take any number of arguments but, unlike normal functions, evaluates and returns only one expression

```python
x = lambda a: a + 10
print(x(10))
```

```python
x = lambda a, b: a*b+10
print(x(2,8))
```

```python
Max = lambda a, b : a if(a > b) else b
print(Max(12,-12))
```

```python
Max = lambda a, b : a if(a > b) else b
print(Max(2,6))
```
