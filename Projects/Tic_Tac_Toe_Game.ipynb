```python
import random

def check_winner(board, player):
  win_conditions = [
      [0,1,2], [3,4,5],[6,7,8], # Rows
      [0,3,6], [1,4,7],[2,5,8], # Columns
      [0,4,8],[2,4,6]           #Diagonals
  ]


  for index in win_conditions:
    if board[index[0]] == board[index[1]] == board[index[2]] == player:
      return True
  return False

def check_draw(board):
  for element in board:
    if element != "X" and element != "O":
      return False
  return True

def display_board(board_list):
  print(" {} | {} | {} "  . format(board_list[0], board_list[1], board_list[2]))
  print("--------------")
  print(" {} | {} | {}"   . format(board_list[3], board_list[4], board_list[5]))
  print("--------------")
  print(" {} | {} | {}"   . format(board_list[6], board_list[7], board_list[8]))
  print("--------------")
  print("              ")

def get_valid_input(available):
  while True:
    position_to_replace = int(input("Please choose a position between 1 and 9:-")) -1
    if position_to_replace >= 0 and position_to_replace < 9 and available[position_to_replace] in range(1,10):
      return position_to_replace
    print("Invalid Input")

def get_computer_choice(available):
  valid_positions = []
  position = 0

  while position < len(available):
    val = available[position]
    if val != 'X' and val != 'O':
      valid_positions.append(position)
    position += 1
  chosen_position = random.choice(valid_positions)
  return  chosen_position

def play_game():
  print("Welcome to Tic-Tac-Toe!")
  player_name = input('Plese enter your name:-')

  while True:
    available = [num for num in range (1,10)]
    display_board(available)

    player_sign = input("Hi, {}. Choose X or O:" .format(player_name)).upper()
    if player_sign == 'O':
      computer_sign = 'X'
      current_player = player_sign
    else:
      computer_sign = "O"
      current_player = player_sign

    while True:
      if current_player == player_sign:
        position_to_replace = get_valid_input(available)
      else:
        position_to_replace = get_computer_choice(available)

      available[position_to_replace] = current_player
      display_board(available)

      if check_winner(available, current_player):
        if current_player == player_sign:
          print("Player", player_name, "wins!")
        else:
          print('Computer wins!')
        break

      if check_draw(available):
        print("It's Draw!")
        break

      if current_player == player_sign:
        current_player = computer_sign
      else:
        current_player = player_sign

    play_again = input("Do you want to play again? (yes/no):")
    if play_again.lower() != 'yes':
      print("Thank you for playing!")
      break

play_game()
```
