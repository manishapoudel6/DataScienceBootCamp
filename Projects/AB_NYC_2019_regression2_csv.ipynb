```python
import pandas as pd
```

```python
df = pd.read_csv('/content/AB_NYC_2019.csv')
df
```

```python
df.info()
```

```python
df['host_name'].dropna()
```

```python
df.dropna(subset=['host_name'], inplace=True)
```

```python
df.dropna(subset=['name'], inplace=True)
```

```python
df.drop(columns=['last_review'], inplace=True)
```

```python
df['reviews_per_month'].fillna(0, inplace=True)
```

```python
df.isna().sum()
```

```python
df
```

```python
df.shape
```

```python
df['reviews_per_month'] = df.apply(lambda row: row['reviews_per_month'] if row['number_of_reviews'] > 0 else None, axis=1)

zero_reviews_properties = df[df['number_of_reviews']> 0]
host_id = zero_reviews_properties['host_id'].iloc[0]
mean_reviews_per_month = df[df['host_id'] == host_id]['reviews_per_month'].mean()
df['reviews_per_month'].fillna(mean_reviews_per_month, inplace=True)
df
```

```python
df.isna().sum()
```

```python
host_id
```

```python
X=df.drop(columns=['id','name','host_id','host_name','price','neighbourhood','availability_365'],axis=1)
y=df['price']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0,3])], remainder='passthrough')
X1 = ct.fit_transform(X)
X1
```

```python
from sklearn.model_selection import train_test_split
X1_train,X1_test,y_train,y_test=train_test_split(X1,y,test_size=0.2,random_state=1)
```

```python
X.isna().sum()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X1_train,y_train)
```

```python
print(reg.coef_,reg.intercept_)
```

```python
prediction = reg.predict(X1_test)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test,prediction)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.xlim(0,400)
```

Using Polynomial Regressor

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 4)
X_train_poly = poly.fit_transform(X1_train)
X_test_poly = poly.transform(X1_test)
```

```python
#Fitting a polynomial regression model to your training data

model = LinearRegression()
model.fit(X_train_poly, X1_train)
```

```python
y_train_pred = model.predict(X_train_poly)
y_test_pred = model.predict(X_test_poly)
```

```python
plt.scatter(X1_train, y_train_pred, label='Testing Data', c='r')
plt.scatter(X1_test, y_test_pred, label='Polynomial Regression (Testing)', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

USing SVR

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler()
X1_train_scaled = scaler.fit_transform(X1_train)
X1_test_scaled = scaler.fit_transform(X1_test)
```

```python
from sklearn.svm import SVR
svr_regressor = SVR(kernel='linear', C=1.0, gamma='scale')
svr_regressor.fit(X1_train_scaled, y_train)
y_test_pred = svr_regressor.predict(X1_test_scaled)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_test_pred, color='blue', label='SVR Predictions')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', lw=2, label='Perfect Fit')


plt.xlabel('X')
plt.ylabel('y')
plt.title('SVR Model Predictions')
plt.legend()
plt.show()
```

Prediction Using Random Regressor

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
regressor.fit(X1_train, y_train)

prediction = regressor.predict(X1_test)

from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, prediction))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, prediction))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
print('R2 score:', np.sqrt(metrics.r2_score(y_test, prediction)))
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, prediction, color='blue', label='Actual vs. Predicted')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values')
plt.legend()
plt.show()
```

Decison_Tree

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X1, y)
```

```python
from sklearn.metrics import mean_squared_error as mse

max_depths = range(1, 20)
training_error = []
for max_depth in max_depths:
    model_1 = DecisionTreeRegressor(max_depth=max_depth)
    model_1.fit(X1, y)
    training_error.append(mse(y, model_1.predict(X1)))

testing_error = []
for max_depth in max_depths:
    model_2 = DecisionTreeRegressor(max_depth=max_depth)
    model_2.fit(X1_train, y_train)
    testing_error.append(mse(y_test, model_2.predict(X1_test)))

plt.plot(max_depths, training_error, color='blue', label='Training error')
plt.plot(max_depths, testing_error, color='green', label='Testing error')
plt.xlabel('Tree depth')
plt.axvline(x=7, color='orange', linestyle='--')
plt.annotate('optimum = 7', xy=(7.5, 1.17), color='red')
plt.ylabel('Mean squared error')
plt.title('Hyperparameter Tuning', pad=15, size=15)
plt.legend()
plt.savefig('error.png')
```
