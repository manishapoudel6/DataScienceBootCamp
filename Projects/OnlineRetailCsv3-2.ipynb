# 2) Create a Dictionary with the 10 countries that have the most 'Quantity' ordered except UK

```python
excluding_uk = df_cleaned[df_cleaned['Country'] != 'United Kingdom']
country_quantities = excluding_uk.groupby('Country')['Quantity'].sum()
top_10_countries = country_quantities.sort_values(ascending=False).head(10)
top_10_countries_dict = top_10_countries.to_dict()
print(top_10_countries_dict)
```

 # 4. Find each Countries Total SALE to find which Countries is highest potential customer do a bar chat for this

```python
countries = df_cleaned.groupby(['Country']).sum()['Total_Sale'].sort_values(ascending=False)
```

```python
country_sales = df_cleaned.groupby('Country')['Quantity', 'UnitPrice'].sum()
country_sales['TotalSale'] = country_sales['Quantity'] * country_sales['UnitPrice']

highest_total_sale = country_sales['TotalSale'].max()

highest_potential_country = country_sales['TotalSale'].idxmax()

print("Total Sale by Country:")
print(country_sales)
print("Highest Potential Customer Country:", highest_potential_country)
print("Highest Total Sale:", highest_total_sale)
```

```python
from pandas.core.frame import Index
import numpy as np

number_of_country=np.arange(len(countries))
fig=plt.figure(figsize=(10,5))
ax=fig.add_axes([0.2,0.2,0.8,0.9])
ax.barh(number_of_country,countries,0.8)
plt.xticks(rotation='vertical')
ax.set_yticks(number_of_country)
ax.set_yticklabels(countries.index.values, minor=False)
plt.xscale('log')
```

#5. Which Customer Id has highest purchase in quantity and total and find out the country of it

```python
customer_sales = df_cleaned.groupby('CustomerID').agg({'Quantity': 'sum', 'UnitPrice': 'sum'})
customer_sales['TotalSale'] = customer_sales['Quantity'] * customer_sales['UnitPrice']

highest_quantity_customer = customer_sales['Quantity'].idxmax()
highest_total_sale_customer = customer_sales['TotalSale'].idxmax()

country_of_highest_total_sale_customer = df_cleaned[df_cleaned['CustomerID'] == highest_total_sale_customer]['Country'].iloc[0]

print("Customer with Highest Purchase Quantity (Customer ID):", highest_quantity_customer)
print("Customer with Highest Total Purchase (Customer ID):", highest_total_sale_customer)
print("Country of Customer with Highest Total Purchase:", country_of_highest_total_sale_customer)
```

#6. Which Customer Id has lowest purchase in quantity and total and find out the country of it

```python
customer_sales = df_cleaned.groupby('CustomerID').agg({'Quantity': 'sum', 'UnitPrice': 'sum'})
customer_sales['TotalSale'] = customer_sales['Quantity'] * customer_sales['UnitPrice']

lowest_quantity_customer = customer_sales['Quantity'].idxmin()
lowest_total_sale_customer = customer_sales['TotalSale'].idxmin()

country_of_lowest_total_sale_customer = df_cleaned[df_cleaned['CustomerID'] == lowest_total_sale_customer]['Country'].iloc[0]

print("Customer with Lowest Purchase Quantity:", lowest_quantity_customer)
print("Customer with Lowest Total Purchase:", lowest_total_sale_customer)
print("Country of Customer with Lowest Total Purchase:", country_of_lowest_total_sale_customer)
```

#7. Find the biggest order In Sale (Cost) which country was the order from and find total purchase (Cost) done by that customer ID

```python
customer_sales = df_cleaned.groupby('CustomerID').agg({'Quantity': 'sum', 'UnitPrice': 'sum'})
customer_sales['TotalSale'] = customer_sales['Quantity'] * customer_sales['UnitPrice']
biggest_order= customer_sales[customer_sales['TotalSale'] == customer_sales['TotalSale'].max()]

customer_id_of_biggest_order = biggest_order.index[0]

country_of_biggest_order = df_cleaned[df_cleaned['CustomerID'] == customer_id_of_biggest_order]['Country'].iloc[0]
total_purchase_cost_by_customer = customer_sales.loc[customer_id_of_biggest_order, 'TotalSale']

print('Customer ID of biggest order:', customer_id_of_biggest_order)
print('Country of biggest order:', country_of_biggest_order)
print('Total purchase cost by customer:', total_purchase_cost_by_customer)
```
