```python
directors = {} #Counting how many movies these each director has made.
def per_director(Name):
    directors[Name] = len(df[df['Director'] == Name])

for value in df['Director'].values:
    per_director(value)

index_labels = ['movies_count']
director = pd.DataFrame(directors, index=index_labels)
director
```
