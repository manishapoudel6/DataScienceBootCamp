```python
#1) School Management System: Design classes for School, Student, and Teacher.
# Students can enroll in courses, teachers can assign grades, and the school can generate reports.

class School:
  def __init__(self, name):
    self.name = name
    self.courses = []
    self.students = []
    self.teachers = []
    self.grades = []


  def add_course(self, course):
    self.courses.append(course)

  def add_student(self, student):
    self.students.append(student)

  def add_teacher(self, teacher):
    self.teachers.append(teacher)

  def generate_school_report(self):
    print(f"School Report for {self.name}:")
    print("Courses:")
    for course in self.courses:
      print(f"- {course.name}")
    print("Students:")
    for student in self.students:
      print(f"- {student.name}, ID: {student.student_id}")
    print("Teachers:")
    for teacher in self.teachers:
      print(f"- {teacher.name}, ID: {teacher.teacher_id}")
    for student in self.students:
      for course in self.courses:
        if course.name in student.grade:
          print(f"- {student.name}'s {course.name} grade is: {student.grade[course.name]}")



class Course:
  def __init__(self, name):
    self.name = name
    self.students_enrolled = []
    self.teacher = []

  def assign_teacher(self, teacher):
    self.teacher = teacher

  def enroll_student(self, student):
    self.students_enrolled.append(student)




class Student:
  def __init__(self, name, student_id):
    self.name = name
    self.student_id = student_id
    self.grade ={}

  def receive_grade(self, course, grade):
    self.grade[course.name] = grade



class Teacher:
  def __init__(self, name, teacher_id):
    self.name = name
    self.teacher_id = teacher_id


  def assign_grade(self, student, course, grade):
    student.receive_grade(course, grade)
```

```python
#School
my_school = School("Bright Future School")

#Course
course1 = Course("Mathematics")
course2 = Course("Science")

#Student
student1 = Student("Sophie", "S101")
student2 = Student("Aria", "S102")

#Teacher
teacher1 = Teacher("Mr. Netra", "T201")
teacher2 = Teacher("Ms.Suman", "T202")

# Generating school report
my_school.add_course(course1)
my_school.add_course(course2)
my_school.add_student(student1)
my_school.add_student(student2)
my_school.add_teacher(teacher1)
my_school.add_teacher(teacher2)

# Assigning grades
teacher1.assign_grade(student1, course1, 95)
teacher2.assign_grade(student1, course2, 88)
teacher2.assign_grade(student2, course2, 92)

my_school.generate_school_report()
```

```python
# 2) Online Quiz Application: Develop classes for Question, Quiz, and User.
# Users can take quizzes, answer questions, and see their scores at the end


# Online Quiz Application: Develop classes for Question, Quiz, and User.
# Users can take quizzes, answer questions, and see their scores at the end.


class Question:
  def __init__(self, text, choices, correct_choice):
    self.text = text
    self.choices = choices
    self.correct_choice = correct_choice

  def is_correct(self, choice):
    return choice == self.correct_choice


class Quiz:
  def __init__(self, name):
    self.name = name
    self.questions = []

  def add_question(self, question):
    self.questions.append(question)

  def conduct_quiz(self):
    print(f"Welcome to the {self.name} quiz!\n")
    score = 0

    for question in self.questions:
      print(question.text)


      for i in range(len(question.choices)):
        choice = question.choices[i]
        print(f"{i + 1}. {choice}")

      user_choice = int(input("Enter your choice (1, 2, 3, ...): "))

      if question.is_correct(user_choice):
        print("Correct!\n")
        score += 1
      else:
        print("Incorrect.\n")


    print(f"Quiz complete! Your score: {score}/{len(self.questions)}")

class User:
    def __init__(self, name):
        self.name = name
        self.scores = {}

    def take_quiz(self, quiz):
        print(f"{self.name}, you are taking the {quiz.name} quiz.\n")
        quiz.conduct_quiz()
        self.scores[quiz.name] = quiz.questions



if __name__ == "__main__":
    question1 = Question("What is the capital of USA?", ["Washington DC", "Caloifornia", "New York"], 1)
    question2 = Question("Which planet has the most Moon?", ["Saturn", "Venus", "Jupiter"], 1)
    question3 = Question("What is 2 * 2?", ["3", "4", "5"], 2)

    quiz = Quiz("General Knowledge Quiz")
    quiz.add_question(question1)
    quiz.add_question(question2)
    quiz.add_question(question3)


    user = User("Sophie")
    user.take_quiz(quiz)
```

```python

#3) Vehicle Rental System: Build classes for Vehicle, Car, and Bike.
#Implement a rental system where customers can rent vehicles and the system can track availability.

class Vehicle:
  def __init__(self, availability=True):
    self.available = availability

  def rent(self):
    if self.available:
      print('Vehicle rented successfully!')
      self.available = False
    else:
      print('Vehicle is not available')

  def return_vehicle(self):
    if not self.available:
      print('Vehicle returned')
      self.available = True


class Car(Vehicle):
  def car_details(self):
    print('This is the availability of the car:', self.available)


class Bike(Vehicle):
  def bike_details(self):
    print('This is the availability of the bike:', self.available)


class RentalSystem:
  def __init__(self):
    self.vehicles = []

  def add_vehicle(self, vehicle):
    self.vehicles.append(vehicle)

  def list_available_vehicles(self):
    available_vehicles = []
    for vehicle in self.vehicles:
      if vehicle.available:
        available_vehicles.append(vehicle)

    if available_vehicles:
      print("Available vehicles:")
      for vehicle in available_vehicles:
        print(f" - {type(vehicle).__name__}")
    else:
        print("No vehicles are available.")


  def rent_vehicle(self, vehicle):
    vehicle.rent()

  def return_vehicle(self, vehicle):
    vehicle.return_vehicle()



car1 = Car()
bike1 = Bike()
rental_system = RentalSystem()

rental_system.add_vehicle(car1)
rental_system.add_vehicle(bike1)
rental_system.list_available_vehicles()


car1.car_details()
bike1.bike_details()

rental_system.rent_vehicle(car1)
rental_system.rent_vehicle(bike1)

rental_system.list_available_vehicles()

rental_system.return_vehicle(car1)
rental_system.return_vehicle(bike1)

rental_system.list_available_vehicles()
```

```python
#4) Social Media System: Create classes for User, Post, and Comment.
#Users can create posts, like posts, and comment on posts. Implement methods to display posts with their interactions.

class User:
  def __init__(self, username):
    self.username = username
    self.posts = []

  def create_post(self, content):
    post = Post(content, self)
    self.posts.append(post)
    return post

  def like_post(self, post):
    post.add_like(self)
    for like in post.likes:
      print(f"{self.username} Liked: ")
      print(f"Total Likes: {len(post.likes)}")


  def comment_post(self, post, text):
    post.add_comment(self, text)
    for comment in post.comments:
      print(f"- {comment.user.username } commented: {comment.text}")

class Post:
  def __init__(self, content, author):
    self.content = content
    self.author = author
    self.likes = []
    self.comments = []
    print("Post Author: {}".format(self.author.username))
    print("Post Content: {}".format(self.content))

  def add_like(self, user):
    self.likes.append(user)


  def add_comment(self, user, text):
    comment = Comment(user, text)
    self.comments.append(comment)


class Comment:
  def __init__(self, user, text):
    self.user = user
    self.text = text
```
